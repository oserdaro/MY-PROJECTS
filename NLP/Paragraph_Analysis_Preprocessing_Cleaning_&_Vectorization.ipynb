## Reading the text file

## Preprocessing: Cleaning & Vectorization

## Cleaning

### Tokenizing

```python
#Tokenize the text by sentences :
sentences = sent_tokenize(text)

#How many sentences are there? :
print (len(sentences))

#Print the sentences :
#print(sentences)
sentences
```

```python
#Tokenize the text with words :
words = word_tokenize(text)

#How many words are there? :
print (len(words))
print("\n")

#Print words :
print (words)
```

### Frequencies

```python
#Import required libraries :
from nltk.probability import FreqDist

#Find the frequency :
fdist = FreqDist(words)

#Print all/10 most common words :
fdist.most_common()
```

```python
#Plot the graph for fdist :
import matplotlib.pyplot as plt

fdist.plot(10)
```

### Removing the punctuation marks

```python
#Empty list to store words:
words_no_punc = []

#Removing punctuation marks :
for w in words:
    if w.isalpha():
        words_no_punc.append(w.lower())

#Print the words without punctution marks :
print (words_no_punc)

print ("\n")

#Length :
print (len(words_no_punc))
```

```python
#Frequency distribution :
fdist = FreqDist(words_no_punc)

fdist.most_common(10)
```

```python
#Plot the most common words on grpah:

fdist.plot(10)
```

### Removing the stop words

```python
from nltk.corpus import stopwords

#List of stopwords
stopwords = stopwords.words("english")
print(stopwords)
```

```python
#Empty list to store clean words :
clean_words = []

for w in words_no_punc:
    if w not in stopwords:
        clean_words.append(w)

print(clean_words)
print("\n")
print(len(clean_words))
```

```python
#Frequency distribution :
fdist = FreqDist(clean_words)

fdist.most_common(10)
```

```python
#Plot the most common words on grpah:

fdist.plot(10)
```

### Tagging

```python
#PoS tagging :
tagged_words = nltk.pos_tag(clean_words)
tagged_words
```

### Extracting Noun Phrases

```python
#Extracting Noun Phrase from text :

# ? - optional character
# * - 0 or more repetations
grammar = "NP : {<DT>?<JJ>*<NN>} "
import matplotlib.pyplot as plt
#Creating a parser :
parser = nltk.RegexpParser(grammar)

#Parsing text :
output = parser.parse(tagged_words)
print (output)

#To visualize :
#output.draw()
```

### Chinking

##### Chunking is extracting meaningful phrases or chunks from a sentence based on its part-of-speech (POS) tags, grouping words together to form larger units, such as noun phrases (NP) or verb phrases (VP). Chinking is the opposite of chunking. It is defining patterns or words that cannot be part of a chunk, specifying what to exclude intentionally from a chunk.


```python
# Chinking example :
# * - 0 or more repetations
# + - 1 or more repetations

#Here we are taking the whole string and then excluding adjectives from that chunk.

grammar = r""" NP: {<.*>+}
               }<JJ>+{"""

#Creating parser :
parser = nltk.RegexpParser(grammar)

#parsing string :
output = parser.parse(tagged_words)
print(output)

#To visualize :
output.draw()
```

### Names Entity Recognition

```python
# NER: Named Entity Recognition :
N_E_R = nltk.ne_chunk(tagged_words,binary=False)  # <<<< False / True
print(N_E_R)

#To visualize :
N_E_R.draw()
```

```python
#Finding synonyms :

#Empty list to store synonyms :
synonyms = []

for words in wordnet.synsets(clean_words[2]):
    for lemma in words.lemmas():
        synonyms.append(lemma.name())

synonyms
```

```python
#Finding antonyms :

#Empty list to store antonyms :
antonyms = []

for words in wordnet.synsets(clean_words[2]):
    for lemma in words.lemmas():
        if lemma.antonyms():
            antonyms.append(lemma.antonyms()[0].name())

#Print antonyms :
antonyms
```
