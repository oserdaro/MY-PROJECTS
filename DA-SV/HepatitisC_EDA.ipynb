```python
df[pd.to_numeric(df['albumin'], errors='coerce').isna()]
```

```python
# Convert the column to numeric values, assign NaN in case of error

for i in ['albumin', 'alt', 'ast' ,'bil' ,'che' ,'chol' ,'crea' ,'ggt']:
    display(df.loc[pd.to_numeric(df[i], errors='coerce').isna()])
    print('-'*95)
```

 The errors='coerce' parameter allows the pd.to_numeric function to replace unconvertible values (for example, text values) with NaN (Not a Number) when trying to convert a column to numeric values. 
 This means that in case of an error you can cleanly convert the column to numeric values
